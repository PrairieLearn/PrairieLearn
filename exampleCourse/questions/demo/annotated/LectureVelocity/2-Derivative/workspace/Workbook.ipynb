Initialize some of the libraries used in this notebook by "running" the cell below.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

The code snippet below plots the function $f(x) = 0.5 x^2 - 2$ in the interval $[-4,7]$, and the secant line betwen points $a = -2$ and $b = 4$.

```python
def f(x):
    return 0.5*x**2 - 2

# Plot the function
a0 = -4
b0 = 7
x = np.linspace(a0, b0)
plt.plot(x,f(x))
plt.grid()
plt.xlim(a0,b0)
plt.ylim(-4,12)

# Plot the secant line
a = -2
b = 4
xmesh = np.linspace(a0, b0, 100)
slope = (f(b)-f(a))/(b-a)
plt.plot(a, f(a),'or')
plt.plot(b, f(b),'or')
plt.plot(xmesh, f(a) + slope*(xmesh-a))
```

We will now start moving the point $b$ closer to point $a$. You can think of point $b$ defined as $b = a + h$. Initially, we set $h = 6$. We will now decrease the value of $h$ so that it approaches zero. 

Let's observe what happens to the secant line when we decrease the value of $h$. In the code snippet below, change the value of h, and run the cell. See what happens. For example, you can use the values $6, 5, 4, ..., 1, 0.5, 0.1$.

```python
## change the value for h below:
h = 6

# do not change the following code
bnew = a + h
slope = (f(bnew)-f(a))/(bnew-a)
plt.plot(x,f(x))
plt.grid()
plt.xlim(a0,b0)
plt.ylim(-4,12)
plt.plot(a, f(a),'or')
plt.plot(bnew, f(bnew),'or')
plt.plot(xmesh, f(a) + slope*(xmesh-a))
```

Interesting right? 

**In the limit when the secant line approaches zero, the secant line approaches the tangent line.**

Repeat the same change in $h$ for the code snippet below. In addition to the secant line, we are also plotting the tangent line.


```python
## change the value for h below:
h = 6

# do not change the following code
bnew = a + h
slope = (f(bnew)-f(a))/(bnew-a)
plt.plot(x,f(x))
plt.grid()
plt.xlim(a0,b0)
plt.ylim(-4,12)
plt.plot(a, f(a),'or')
plt.plot(bnew, f(bnew),'or')
plt.plot(xmesh, f(a) + slope*(xmesh-a))
plt.plot(xmesh, f(a) + a*(xmesh-a))
```
